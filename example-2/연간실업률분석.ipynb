```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
sns.set(style="whitegrid")
```

```python
import matplotlib.font_manager as fm
import platform

if platform.system() =='Windows':
    font_path = 'C:/Windows/Fonts/gulim.ttc' #윈도우 굴림폰트
elif platform.system() == 'Linux':    #우분투 18.04 테스트됨, 나눔폰트 설치 후
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
    
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 12
```

```python
data1 = pd.read_excel('./연간고용률01.xlsx', encoding='utf-8')
data1
```

```python
import matplotlib.ticker as ticker
plt.figure(figsize=(15,5))
ax = sns.lineplot(x="연도", y='고용율', marker=True,sizes=19, markers="o", data=data1)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.show()
```

```python
x = data1.연도.tolist()
y = data1.고용율.tolist()
plt.figure(figsize=(15,4))
ax = plt.plot(x,y,'ro-', ms=8, mfc='g')
plt.title('연도별 고용율 변화')
plt.xlabel('연도')
plt.ylabel('고용율')
plt.ylim(57.5, 62.5)
plt.xticks(x)
plt.show()
```

```python

data2 = pd.read_excel('./2000-18_실업률_고용률.xlsx', encoding='utf-8')
data2
x = data2.index.tolist() #연도추출
y = data2.실업률.tolist() #실업률
plt.figure(figsize=(15,4))
ax = plt.plot(x,y, 'mo-', ms=8, mfc='r')
plt.xticks(x)
plt.title('연도별 전체 실업률 변화')
plt.xlabel('연도(년)')
plt.ylabel('실업률(%)')
plt.show()
```

```python
data3 = pd.read_excel('./실업률_1990_2018_ver2.xlsx', encoding='utf-8')
data3
```

```python
data3.set_index('Unnamed: 0', drop=True, append=False, inplace=True)
```

```python
data3
```

```python
#전체실업률
x = data3.columns
y = data3.loc['실업률'].tolist()
plt.figure(figsize=(15,4))
ax = plt.plot(x,y, 'mo-', ms=8, mfc='r')
plt.xticks(x, rotation=45)
plt.title('연도별 전체 실업률 변화')
plt.xlabel('연도(년))')
plt.ylabel('실업율(%)')
plt.show()
```

```python
#청년실업률
x = data3.columns
y = data3.loc['청년실업률'].tolist()
plt.figure(figsize=(15,4))
ax = plt.plot(x,y, 'mo-', ms=10, mfc='r')
plt.xticks(x, rotation=45)
plt.title('청년 실업률 변화')
plt.xlabel('연도(년))')
plt.ylabel('실업율(%)')
plt.show()
```

```python
x1 = data3.columns
y1 = data3.loc['실업률'].tolist()

x2 = data3.columns
y2 = data3.loc['청년실업률'].tolist()
plt.figure(figsize=(15,4))
ax1 = plt.plot(x1,y1, 'ro-', ms=5, label='전체 실업률')
ax2 = plt.plot(x2,y2, 'bo--', ms=5, label='청년 실업률')
plt.xticks(x, rotation=45)
plt.title('실업률 비교')
plt.xlabel('연도(년))')
plt.ylabel('실업율(%)')
plt.legend()
plt.show()
```

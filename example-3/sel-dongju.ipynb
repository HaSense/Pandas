### 문자열값 정수로

```python
import pandas as pd

df = pd.DataFrame({'A': ['0', '1', '2', '3'], 
                    'B': ['4', '5', '6', '7'],
                    'C': ['8', '9', '10', '11'],
                    'D': ['12', '13', '14', '15']},
                   index=[0, 1, 2, 3])
df
```

```python
df.A.values
```

```python
df['A'] = df.A.astype('int64')
```

```python
df.dtypes
```

```python
df['B']=df.B.astype('int64')
```

```python
df = df.astype({'C':'int64','D':'int64'})
```

```python
edgeilab = folium.Map(location=[36.334477, 127.457733], tiles='OpenStreetMap',zoom_start=16)

folium.Marker([36.334477, 127.457733], popup='엣지아이랩').add_to(edgeilab)
folium.CircleMarker([36.334477, 127.457733], radius=70, popup='대전대 주변반경', color="#3186cc", fill_color="#424200").add_to(edgeilab)

edgeilab
```

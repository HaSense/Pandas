```python
import matplotlib.pyplot as plt
```

```python
#코드에 한글이 들어가는 것은 일반적으로 좋지 않습니다.
#단, 데이터분석에 한해서는 동작에 오류가 적고 직관적인 인식 및 코드수정에 유용하여 권장하지 않지만 금지하지 않습니다.
data1.이름
```

```python
#원하는 형태로 가공하기 위해 데이터의 타입을 변환해야 합니다.
name = data1.이름.tolist()  #파이썬의 리스트 타입으로 만들어주는 함수
name
```

```python
#잡음발생!!! 데이터의 첫자리에 공백이 삽입됨 제거를 합시다.
#데이터 정제
name2=[]  # 빈 변수를 생성
for i in name:
    name2.append(i.strip()) #좌우공백제거

name2
```

```python
#데이터프레임소스 값 변경
data1.이름 = name2
#data1.이름    #타입의 모양을 테스트해 보세요.
data1.이름.tolist()
```

matplotlib에서 한글폰트 사용하려면~!!!

```python
import matplotlib.font_manager as fm
import platform

if platform.system() =='Windows':
    font_path = 'C:/Windows/Fonts/gulim.ttc' #윈도우 굴림폰트
elif platform.system() == 'Linux':    #우분투 18.04 테스트됨, 나눔폰트 설치 후
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
    
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 12
```

```python
plt.figure(figsize=(10,4))
plt.plot(name, hr)
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(name, hr)
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.bar(name, hr, color='#0000ff')
plt.xlabel('선수이름')
plt.xticks(rotation=45)
plt.ylabel('홈런수')
plt.title('2018년 프로야구 홈런타자 순위')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.pie(hr, labels=name, autopct='%1.2f %%', shadow=False, startangle=-20)
plt.title('2018년 프로야구 홈런타자 순위')
plt.show()
```

```python
import seaborn as sns
 
sns.set(style="darkgrid")
```

```python
#sns.set()을 실행하면 한글 설정이 날아가게 됩니다. <br>
#다시 한글설정 모듈을 삽입합니다.

import matplotlib.font_manager as fm
import platform

if platform.system() =='Windows':
    font_path = 'C:/Windows/Fonts/gulim.ttc' #윈도우 굴림폰트
elif platform.system() == 'Linux':    #우분투 18.04 테스트됨, 나눔폰트 설치 후
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
    
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 12
```

```python
#seaborn
plt.figure(figsize=(15,4))
chart = sns.barplot(x="이름", y="홈런", data=data1) #seaborn은 데이터프레임을 직접 다루는 함수가 많습니다.
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
plt.show()
```
